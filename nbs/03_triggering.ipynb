```python
#| hide
from nbdev.showdoc import *
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
#|hide
sns.set()
sns.set_context('talk')
sns.set_style('ticks')
```

## Start signal

```python
#|echo:false
# Load data
data = np.loadtxt('./example_data/start_trigger.csv', delimiter=',', skiprows=2, usecols=[0,1])

t = data[:, 0]
v = data[:, 1]

t = t + 0.5

fig, ax = plt.subplots()

ax.plot(t, v, 'k')

ax.set(
    yticks = (0, np.max(v)),
    yticklabels = ['Off', 'On'],
    xlim = (0, 2),
    xlabel = 'Time (s)'
)

sns.despine(ax=ax, offset=10, trim=True)
```
